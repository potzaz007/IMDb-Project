```python
#โจทย์คือหาจำนวนหนัง Action และจำนวนนักแสดง Actiom
```

```python
import pandas as pd
        
df = pd.read_csv('IMDb movies .csv')
df
```

```python
df.info()
```

```python
#หาจำนวนหนัง Action
```

```python
df['Action_genre'] = df['genre'].apply(lambda member: 'Action' in member)
Action_movie = df[df['Action_genre']]
Action_movie
```

```python
#เลือก ID นักแสดง Action 
```

```python
import pandas as pd
        
IMDb_title_principals = pd.read_csv('IMDb title_principals.csv')
IMDb_title_principals
```

```python
#รวมตารางโดยใช้คอลัมน์ที่เหมือนกัน imdb_title_id
```

```python
merged_df = Action_movie.merge(IMDb_title_principals, how='inner', left_on='imdb_title_id', right_on='imdb_title_id')
merged_df
```

```python
#กรองเฉพาะที่เป็นนักแสดงผลลัพธ์ที่ได้จะได้นชื่อนักแสดงที่อาจซ้ำกันได้จึงต้องเปรียบเทียบกับ Name ID
```

```python
action_star_df = merged_df[merged_df['category'].isin(['actor','actress'])]
action_star_df
```

```python
#อ่านไฟล์ Name ID
```

```python
import pandas as pd
        
IMDb_names = pd.read_csv('IMDb names.csv')
IMDb_names
```

```python
#รวตารางชื่อนักแสดง Action และ Name ID จากนั้นใช้คำสั่ง drop_duplicates() กรองชื่อซ้ำ
```

```python
action_star_df.merge(IMDb_names, how='inner', left_on='imdb_name_id', right_on='imdb_name_id')[['imdb_name_id','name']].drop_duplicates()
```
